

>En el diseño de un sistema de cinta transportadora, pequeños bloques de metal de $5kg$ se descargan con una velocidad de

Con el valor del ángulo obtenido, realice un gráfico del trabajo realizado por el peso y la fricción en función de la altura. ¿Qué conclusiones puede tener con respecto al trabajo en este caso?


>Las fuerzas actuando sobre las caja son la normal, fuerza de rozamiento y peso. Por tanto, analizando los puntos 1 y 2, y aplicando el principio de trabajo y energía cinética se obtiene:


$$T_1+U_{1-2}=T_2$$

$$\frac{1}{2}m{V_1}^2+mgh-U_{f_r}=\frac{1}{2}m{V_2}^2$$

$$\frac{1}{2}m{V_1}^2+mgh-{f_r}\frac{15}{sen(\theta)}=\frac{1}{2}m{V_2}^2$$


>Para hallar el valor de la fuerza de rozamiento, se realiza un DCL de una caja, colocando un sistema de referencia paralelo al plano inclinado, de esta forma se obtiene:

$$\sum{{F_{y'}}}=0$$

$$Pcos(\theta)=N$$

$$N=mgcos(\theta)$$

>Considerando que $f_r=\mu N$ 
y reemplzando el valor de la normal, se obtiene:

$$\frac{1}{2}m{V_1}^2+mgh-\mu  mgcos(\theta)\frac{1.5}{sen(\theta)}=\frac{1}{2}m{V_2}^2$$

$$1,5g\mu cot(\theta)=\frac{1}{2}{V_{1}}^2+gh-\frac{1}{2}{V_{2}}^2$$

$$cot(\theta)=3,349$$

$$\theta=16,62°$$


>Para el caso del trabajo de la fricción en función de la distancia recorrida, se tiene:
$$sen(\theta)=\frac{h}{dr}$$
$$dr=\frac{h}{sen(\theta)}$$
$$\frac{1}{2}m{V_1}^2+mgh-\mu *mgcos(\theta)\frac{h}{sen(\theta)}=\frac{1}{2}m{V_2}^2$$
Hay dos fuerzas que realizan trabajo, que son el peso y la fricción. 
Entonces el trabajo realizado por estas fuerzas, en función de la altura, está dado por:
$$U_{1-2}(h)=mgh-\mu mg\frac{cos(\theta}{sen(\theta)}h$$
$$U_{1-2}(h)=5(9.8)h-0.3(5)(9.8)(3.35016)h$$
$$U_{1-2}(h)=\left(5(9.8)-0.3(5)(9.8)(3.35016)\right)h$$



```python
import plotly.express as px
import numpy as np

h=np.arange(0,1.50,0.01)
W=(5*9.81*h)-(0.3*5*9.8*3.35016*h)
#plot
fig=px.line(x=h,y=W,labels={'x':'altura ($m$)','y':'Trabajo $(Nm)$'},title="Trabajo en funcion de la altura")
fig.show()
```

>Como observamos en el gráfico el trabajo es resitivo desde el punto $1$ hasta el punto $2$ y esto se puede comprobar debido a que posee menor velocidad en dicho punto en relación al inicio. Se puede decir que esto se debe al trabajo de la fricción es mayor al trabajo del eso puesto que en este caso el trabajo del peso es positivo por tener el desplazamiento en la misma dirección de la fuerza. Para el caso de la fricción se tiene que el trabajo es resistivo y supera al trabajo del peso por lo que disminuye la energía cinética de las cajas.
